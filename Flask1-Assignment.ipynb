```python
Q1. What is Flask Framework? What are the advantages of Flask Framework?
```

```python

Flask is a micro web framework written in Python. It is lightweight and designed to be simple yet powerful, 
allowing developers to build web applications quickly and efficiently. Flask provides tools, libraries, and features
for creating web applications, APIs, and more.

Advantages of Flask Framework include:

Lightweight: 
    Flask is minimalistic and does not come with many built-in features, which makes it lightweight and flexible.
    Developers can choose the components they need and add them as necessary.
    
Flexibility: 
    Flask allows developers to structure their applications as they see fit. It does not impose any specific project 
    layout, giving developers the freedom to organize their code according to their preferences.
    
Easy to Learn:
    Flask has a simple and intuitive syntax, making it easy for beginners to learn and start building web applications.
    Its minimalistic design also helps reduce the learning curve.
    
Extensibility:
    Flask is highly extensible, allowing developers to add functionality through Flask extensions. There are numerous
    extensions available for various purposes, such as authentication, DB integration, and form validation.
    
Built-in Development Server: 
    Flask comes with a built-in development server, which makes it easy to develop and test applications locally
    without needing to set up a separate server environment.
```

```python
In Flask, app routing is the mechanism that establishes a connection between specific URLs (web addresses) and the 
corresponding Python functions that handle requests for those URLs.
When you create a Flask application, you define routes using the @app.route() decorator,
which binds a function to a URL.

We use app routes to organize our applications logic and create a structured navigation system. This makes it easier to
manage different parts of the application and allows users to access specific functionalities or resources by visiting 
different URLs. Also, app routes help in creating RESTful APIs where different routes correspond to different actions
or resources.
```

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, World!'
```

```python
Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.
```

```python
In Flask, the url_for() function is used for URL building.
It generates a URL to a specific function within your Flask application based on its endpoint name.
```

```python
from flask import Flask,url_for

app = Flask(__name__)

@app.route("/welcome")
def hello_world():
    return "WELCOME MY FRIEND"

if __name__=="__main__":
    app.run(host="0.0.0.0")
    with app.test_Request_context():
        print(url_for('hello_world'))
```
